```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import ipywidgets as widgets
from ipywidgets import interact
```

```python
df['All Fields - Female %'] = (df['All Fields - Female']/(df['All Fields - Female'] + df['All Fields - Male']))*100
df['All Fields - Male %'] = (df['All Fields - Male']/(df['All Fields - Female'] + df['All Fields - Male']))*100
df['All Fields - Female %'] = round(df['All Fields - Female %'])
df['All Fields - Male %'] = round(df['All Fields - Male %'])

df['Non-S&E - Female %'] = (df['Non-S&E - Female']/(df['Non-S&E - Female'] + df['Non-S&E - Male']))*100
df['Non-S&E - Male %'] = (df['Non-S&E - Male']/(df['Non-S&E - Female'] + df['Non-S&E - Male']))*100
df['Non-S&E - Female %'] = round(df['Non-S&E - Female %'])
df['Non-S&E - Male %'] = round(df['Non-S&E - Male %'])

df['All S&E - Female %'] = (df['All S&E - Female']/(df['All S&E - Female'] + df['All S&E - Male']))*100
df['All S&E - Male %'] = (df['All S&E - Male']/(df['All S&E - Female'] + df['All S&E - Male']))*100
df['All S&E - Female %'] = round(df['All S&E - Female %'])
df['All S&E - Male %'] = round(df['All S&E - Male %'])

df['Agricultural Sciences - Female %'] = (df['Agricultural Sciences - Female']/(df['Agricultural Sciences - Female'] + df['Agricultural Sciences - Male']))*100
df['Agricultural Sciences - Male %'] = (df['Agricultural Sciences - Male']/(df['Agricultural Sciences - Female'] + df['Agricultural Sciences - Male']))*100
df['Agricultural Sciences - Female %'] = round(df['Agricultural Sciences - Female %'])
df['Agricultural Sciences - Male %'] = round(df['Agricultural Sciences - Male %'])

df['Engineering - Female %'] = (df['Engineering - Female']/(df['Engineering - Female'] + df['Engineering - Male']))*100
df['Engineering - Male %'] = (df['Engineering - Male']/(df['Engineering - Female'] + df['Engineering - Male']))*100
df['Engineering - Female %'] = round(df['Engineering - Female %'])
df['Engineering - Male %'] = round(df['Engineering - Male %'])

df['Mathematics & Computer Sciences - Female %'] = (df['Mathematics & Computer Sciences - Female']/(df['Mathematics & Computer Sciences - Female'] + df['Mathematics & Computer Sciences - Male']))*100
df['Mathematics & Computer Sciences - Male %'] = (df['Mathematics & Computer Sciences - Male']/(df['Mathematics & Computer Sciences - Female'] + df['Mathematics & Computer Sciences - Male']))*100
df['Mathematics & Computer Sciences - Female %'] = round(df['Mathematics & Computer Sciences - Female %'])
df['Mathematics & Computer Sciences - Male %'] = round(df['Mathematics & Computer Sciences - Male %'])

df['Physical & Biological Sciences - Female %'] = (df['Physical & Biological Sciences - Female']/(df['Physical & Biological Sciences - Female'] + df['Physical & Biological Sciences - Male']))*100
df['Physical & Biological Sciences - Male %'] = (df['Physical & Biological Sciences - Male']/(df['Physical & Biological Sciences - Female'] + df['Physical & Biological Sciences - Male']))*100
df['Physical & Biological Sciences - Female %'] = round(df['Physical & Biological Sciences - Female %'])
df['Physical & Biological Sciences - Male %'] = round(df['Physical & Biological Sciences - Male %'])

df['Social & Behavioral Sciences - Female %'] = (df['Social & Behavioral Sciences - Female']/(df['Social & Behavioral Sciences - Female'] + df['Social & Behavioral Sciences - Male']))*100
df['Social & Behavioral Sciences - Male %'] = (df['Social & Behavioral Sciences - Male']/(df['Social & Behavioral Sciences - Female'] + df['Social & Behavioral Sciences - Male']))*100
df['Social & Behavioral Sciences - Female %'] = round(df['Social & Behavioral Sciences - Female %'])
df['Social & Behavioral Sciences - Male %'] = round(df['Social & Behavioral Sciences - Male %'])
```
